```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(xi,a,n):
    x=np.zeros(n)
    x[0]=xi
    for i in range(n-1):
        x[i+1]=a*x[i]*(1-x[i])
    return x
```

```python
def df(xi,a,n):
    x=np.zeros(n)
    y=f(xi,a,n)
    for i in range(n):
        x[i]=a*(1-2*y[i])
    return x
```

```python
la=np.zeros(1000)
a2=np.linspace(3.0,3.999,1000)
for i in range(1000):
    a=3.0+1*i/1000
    la[i]=np.sum(np.log(abs(df(0.2,a,10000))))/10000
plt.figure(figsize=(4,3),dpi=98)
plt.plot(a2,la,linewidth=1)
plt.grid(True)
plt.xlabel('a',fontsize=10)
plt.ylabel('$\lambda$',fontsize=10)
plt.show()
```
