```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(xi,a,n):
    x=np.zeros(n)
    x[0]=xi
    for i in range(n-1):
        x[i+1]=a*np.sin(np.pi*x[i])
    return x
```

```python
n=np.linspace(0,49,50)
plt.figure(figsize=(4,3),dpi=98)
plt.plot(n,f(1/2,1/24,50),linewidth=1)
plt.plot(n,f(1/2,1/4,50),linewidth=1)
plt.plot(n,f(1/2,1/2,50),linewidth=1)
plt.plot(n,f(1/2,3/4,50),linewidth=1)
plt.grid(True)
plt.legend()
plt.xlabel('x',fontsize=10)
plt.ylabel('y',fontsize=10)
plt.show()
```

```python
n=np.linspace(4901,5000,100)
#plt.figure(figsize=(4,3),dpi=98)
plt.plot(n,f(1/2,0.72,5000)[4900:],linewidth=1)
plt.plot(n,f(1/2,0.73,5000)[4900:],linewidth=1)
#plt.plot(n,f(1/2,0.74,5000)[4900:],linewidth=1)
#plt.plot(n,f(1/2,3/4,5000)[4900:],linewidth=1)
plt.grid(True)
plt.legend()
plt.xlabel('x',fontsize=10)
plt.ylabel('y',fontsize=10)
plt.show()
```

Below the $a=0.75$, there are still twofold cycles.

```python
print(f(1/2,3/4,100))
```

Therefore, at $a=3/4$, this is a two fold cycle but not a bifurcation.
