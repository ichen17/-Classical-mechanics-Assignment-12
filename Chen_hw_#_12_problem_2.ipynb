```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def f(xi,a,n):
    x=np.zeros(n)
    x[0]=xi
    for i in range(n-1):
        x[i+1]=a*np.sin(np.pi*x[i])
    return x
```

```python
for i in range(1000):
    a=i/1000
    x=f(0.3,a,5000)
    if(np.abs(x[4997]-x[4998])>1e-7 and  np.abs(x[4997]-x[4999])<1e-9):
        print(a)
        break
```

```python
n=np.linspace(4900,5000,100)
plt.figure(figsize=(4,3),dpi=98)
plt.plot(n,f(0.3,0.721,5000)[4900:],linewidth=1)
plt.grid(True)
plt.legend()
plt.xlabel('x',fontsize=10)
plt.ylabel('y',fontsize=10)
plt.show()
```
